```python
# read data

if not hasattr(body, "__iter__"): body.__iter__ = types.MethodType( __iter__, body )

df1 = pd.read_csv(body)
df1.head()
```

```python
# split features and target

x = df1.iloc[:, :-1]
y = df1.iloc[:, -1:]
x
```

```python
# create random samples of training and testing data

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3)
```

```python
# model fitting

clf = LinearRegression()
clf.fit(x_train, y_train)
```

```python
# prediction

y_pred = clf.predict(x_test)
y_pred
```

```python
# performance score

r2_score(y_pred, y_test) * 100
```

```python
# plot the linear regression model

plt.scatter(x_train, y_train, color='red')
plt.plot(x_train, clf.predict(x_train), color='blue')
plt.title("Salary vs Experience (Based on training set)")
plt.xlabel("Years of Employees")
plt.ylabel("Salary")
plt.show()
```
